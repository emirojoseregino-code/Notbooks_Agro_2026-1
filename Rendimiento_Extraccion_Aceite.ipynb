# Control de Rendimiento en Extracción de Aceite #
Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento**
Una planta procesadora de palma de aceite necesita  calcular.  **el Índice de Rendimiento de Extracción (IRE)**  El operario debe ingresar el **peso total de los racimos de fruta fresca (RFF)** procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

##Requerimientos ##

1. solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float).
3. Solicitar los litros de aceite obtenidos (Tipo float).
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L (r)

$$Peso Aceite (kg)=Litros*0.92$$

$$Rendimiento=(Peso Aceite/Peso Racimos)*100$$

**Prueba del programa**

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%

Aquí está la explicación de cada línea del código:

```python
# Entrada: Ingresar solicitar el nombre del lote de producción (Tipo string).
# solicitar el peso de los racimos en kg (Tipo float). (pr)
# Solicitar los litrosceite obtenidos (Tipo float). (lt)
```
Estas líneas son comentarios que describen los datos de entrada que el programa solicitará al usuario.

```python
nombre = input("ingrese el nombre del lote: ")
```
Esta línea pide al usuario que ingrese el nombre del lote de producción. El texto dentro de los paréntesis `"ingrese el nombre del lote: "` es lo que se muestra al usuario como una solicitud. La entrada del usuario se almacena como una cadena (string) en la variable `nombre`.

```python
pr = float(input("ingrese el peso de los racimos en kg: "))
```
Esta línea solicita al usuario el peso de los racimos en kilogramos. `input("ingrese el peso de los racimos en kg: ")` obtiene la entrada del usuario como una cadena. `float()` convierte esa cadena a un número decimal (float), ya que el peso puede no ser un número entero. El resultado se guarda en la variable `pr` (peso de los racimos).

```python
lt = float(input("ingrese los litros de aceite obtenidos: "))
```
Similar a la línea anterior, esta línea solicita los litros de aceite obtenidos. La entrada del usuario se convierte a un número decimal (float) usando `float()` y se guarda en la variable `lt` (litros de aceite).

```python
# Proceso: calcular peso del aceite (pr) y el rendimiento (r)
```
Este es un comentario que indica que las siguientes líneas realizarán los cálculos.

```python
pa = lt*0.92
```
Esta línea calcula el peso del aceite en kilogramos. Multiplica los `litros` de aceite (`lt`) por su densidad, que es `0.92 kg/L`. El resultado se almacena en la variable `pa` (peso del aceite).

```python
r = (pa/pr)*100
```
Esta línea calcula el rendimiento de extracción. Divide el `peso del aceite` (`pa`) entre el `peso de los racimos` (`pr`) y luego multiplica el resultado por `100` para obtener un porcentaje. El valor se guarda en la variable `r` (rendimiento).

```python
# Salida; informar el rendimiento del lote
print("el rendimiento del lote", nombre, "es del", r, "%")
```
Estas líneas se encargan de la salida del programa. La función `print()` muestra el resultado al usuario, concatenando varias cadenas y variables para formar un mensaje informativo que incluye el nombre del lote y el rendimiento calculado.

````python
# Entrada: Ingresar solicitar el nombre del lote de producción (Tipo string).
# solicitar el peso de los racimos en kg (Tipo float). (pr)
# Solicitar los litrosceite obtenidos (Tipo float). (lt)

nombre = input("ingrese el nombre del lote: ")
pr = float(input("ingrese el peso de los racimos en kg: "))
lt = float(input("ingrese los litros de aceite obtenidos: "))

# Proceso: calcular peso del aceite (pr) y el rendimiento (r)
pa = lt*0.92
r = (pa/pr)*100
# Salida; informar el rendimiento del lote
print("el rendimiento del lote", nombre, "es del", r, "%")


# Salida: informar el rendimiento del lote
````
