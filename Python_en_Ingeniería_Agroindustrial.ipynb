### Ejemplos prácticos para Ingeniería Agroindustrial

Aquí tienes tres scripts sencillos que ilustran cómo Python puede aplicarse en el día a día de un ingeniero:

#### 1. Cálculo del Índice de Madurez

Este código ayuda a determinar si un fruto está listo para la cosecha basándose en la relación entre sólidos solubles (°Brix) y acidez.

```
# Cálculo simple de relación de madurez
grados_brix = 12.5
acidez_titulable = 0.8  # porcentaje

indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Óptimo para cosecha/procesamiento.")
else:
    print("Estado: Aún requiere tiempo en planta.")
```

#### 2. Estimación de Pérdida de Peso por Evapotranspiración

Ideal para monitorear productos en cámaras de almacenamiento postcosecha.

```
# Estimación de pérdida de peso en almacenamiento
peso_inicial = 500.0  # kg
tasa_perdida_diaria = 0.005  # 0.5% por día
dias = 10

peso_final = peso_inicial * (1 - tasa_perdida_diaria)**dias

print(f"Después de {dias} días, el peso estimado es: {peso_final:.2f} kg")
print(f"Pérdida total: {peso_inicial - peso_final:.2f} kg")
```

#### 3. Control de Calidad de Lotes (Promedios)

Un script para automatizar el reporte de calidad de un lote de materia prima que ingresa a la planta.

```
# Registro de diámetros de fruta para clasificación
diametros = [75, 80, 78, 82, 70, 85, 77]  # mm

promedio = sum(diametros) / len(diametros)

print(f"Análisis de Lote:")
print(f"Cantidad de muestras: {len(diametros)}")
print(f"Diámetro promedio: {promedio:.1f} mm")
print(f"Diámetro máximo registrado: {max(diametros)} mm")
```
