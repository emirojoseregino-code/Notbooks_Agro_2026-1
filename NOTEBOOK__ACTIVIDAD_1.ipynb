# Requerimientos del problema: #

Problema: Calcular cuántos bultos de cemento se necesitan para un muro de ladrillo. El usuario ingresa el área del muro en m2
.

Datos: Se estima que por cada 1 m2
 de muro se requieren 0.25
 bultos de cemento.

Prueba: Muro de 40 m2
. Salida: 10 bultos de cemento.

# Prueba del programa: #

40 × 0.25 = 10 bultos


```python
area = float(input("Área del muro: "))

bultos = area * 0.25

print("Bultos necesarios:", bultos)
```
