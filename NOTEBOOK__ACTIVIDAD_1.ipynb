# Prueba del programa: #
Prueba de escritorio

Radio = 3

Altura = 10

Volumen = 3.1416×9×10 = 282.74

Toneladas = 282.74×0.75 ≈ 212.06 toneladas




```python
radio = float(input("Ingrese el radio del silo: "))
altura = float(input("Ingrese la altura del silo: "))

volumen = 3.1416 * radio * radio * altura
toneladas = volumen * 0.75

print("Capacidad del silo:", toneladas, "toneladas")
```
