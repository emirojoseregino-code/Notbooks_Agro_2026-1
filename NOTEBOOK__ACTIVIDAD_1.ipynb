# Requerimientos del problema: #

Problema: Una planta procesadora necesita saber el porcentaje de aprovechamiento.

Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).

Datos:  Pulpa=Peso Total−Residuos - %Rendimiento=(Pulpa/Peso Total)×100


Prueba: Peso Total = 500kg, Residuos = 150kg. Salida: 350kg de pulpa (70%).


# Prueba del programa: #

500 - 150 = 350 kg

(350/500)*100 = 70%


```python
total = float(input("Peso total: "))
residuos = float(input("Residuos: "))

pulpa = total - residuos
rendimiento = (pulpa / total) * 100

print("Pulpa:", pulpa, "kg")
print("Rendimiento:", rendimiento, "%")
```
